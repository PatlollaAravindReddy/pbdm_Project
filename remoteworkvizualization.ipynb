```python
client = bigquery.Client(project='impactofremotework')

query = """
    SELECT *
    FROM `remotework.remotework_temp`
"""

query_job = client.query(query)
df = query_job.to_dataframe()
df.head()
```

```python
# prompt: Age bar vertical

import seaborn as sns
import matplotlib.pyplot as plt

df.groupby('Age').size().plot(kind='bar', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
stress_counts = df['Stress_Level'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(stress_counts, labels=stress_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'orange', 'yellow'])
plt.title('Distribution of Stress Levels', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
grouped = df.groupby(['Work_Location', 'Mental_Health_Condition']).size().unstack()
grouped.plot(kind='bar', figsize=(10, 6))
plt.title('Mental Health Conditions by Work Location')
plt.xlabel('Work Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Mental Health Condition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Years_of_Experience', y='Stress_Level', marker='o')

plt.title('Years of Experience vs. Stress Level')
plt.xlabel('Years of Experience')
plt.ylabel('Stress Level')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Gender', y='Stress_Level', marker='o')

plt.title('Gender vs. Stress Level')
plt.xlabel('Gender')
plt.ylabel('Stress Level')
plt.grid(True)
plt.show()
```
